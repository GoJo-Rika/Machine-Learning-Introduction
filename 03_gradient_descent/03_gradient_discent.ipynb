## Explanation for calculating the optimal cost functiona by taking into account the different learning rates at various iterations

Initially keep the iterations less, say 10, and start with some learning rate, say 0.001 in this example. Now check if the cost function is reducing on each iterations.

After that we can check the answer and tweak the cost function and the learning rate accordingly.

Bigger step(learning rate) = 0.01 with same iterations, it is still reducing.

After that step size of 0.1 with same iterations and the cost function is increasing. So our step size has to be between 0.01 and 0.1. Keeping that in mind and checking for different values of step size and the cost function. We can reach the optimal step size (learning rate) and then change the number of iterations going forward.

## Comparing the values of m & b using sklearn. 

Using math.isclose() with rel_tol (relative tolerance) to check the values with some tolerance.

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def predict_using_sklearn():
    df = pd.DataFrame(np.array([[1, 5], [2,7], [3,9], [4,11], [5,13]]), 
                      columns=['x', 'y'])
    
    reg = LinearRegression()
    reg.fit(df[['x']],df.y)
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(df.x, df.y, color='red', marker='+')
    plt.plot(df.x, reg.predict(df[['x']]), color='blue')
    
    return reg.coef_, reg.intercept_
```

```python
def gradient_descent(x,y):
    x = np.array([1,2,3,4,5])
    y = np.array([5,7,9,11,13])
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    
    cost_previous = 0
    plt.scatter(x,y,color='black',marker='+',linewidth=5)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n)*sum((y - y_predicted)**2)
        
        plt.plot(x,y_predicted, color='green')
        
        md = -(2/n)*sum(x*(y - y_predicted))
        bd = -(2/n)*sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print(f'm {m_curr}, b {b_curr}, cost {cost}, iterations {i}')
    
    return m_curr, b_curr
```

```python
%matplotlib inline
m, b = gradient_descent(x,y)
print(f'Using Gradient Descent Function: Coef {m} and Intercept {b}')

m_sklearn, b_sklearn = predict_using_sklearn()
print(f'Using sklearn: Coef {m_sklearn} and Intercept {b_sklearn}')
```
